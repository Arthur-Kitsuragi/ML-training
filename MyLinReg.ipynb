```python
class llinearRegression:
    def fit(self, features, labels):
        b = np.empty(shape=[features.shape[0], features.shape[1] + 1])
        for i in range (0, features.shape[0]):
            b[i][0] = 1
            for g in range(0, features.shape[1]):
                b[i][g + 1] = features[i][g]
        self.__features = b
        self.__labels = labels
        num1 = np.dot(b.T, b)
        num2 = np.dot(b.T, self.__labels)
        self.__res = np.linalg.solve(num1, num2)
        return self.__res
    def predict(self, data):
        pred = np.empty(shape=[data.shape[0], 1])
        for i in range (0, data.shape[0]):
            sums = self.__res[0]
            for g in range (1, self.__res.shape[0]):
                sums+=self.__res[g] * data[i][g-1]
            pred[i][0] = sums
        return pred
        
```
